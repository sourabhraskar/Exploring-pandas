```python
#  will use groupby command when u want to use it lets say for each that is in our drinks data suppose u want to calculate mean of each continent and continent contains differen groups so u will use groupby
drinks.groupby('continent').beer_servings.mean()        
#this will give mean of examples containing continent as asia
drinks[drinks.continent=='Asia'].beer_servings.mean()   
#this will give mean of examples containing continent as europe
drinks[drinks.continent=='Europe'].beer_servings.mean()  
```

```python
#agg command
#.agg command will give count max min mean of each continent
drinks.groupby('continent').beer_servings.agg('count')
drinks.groupby('continent').beer_servings.agg('max')
drinks.groupby('continent').beer_servings.agg('min')
drinks.groupby('continent').beer_servings.agg('mean')
```
